```python
import pandas as pd
import json
import re
from tqdm import tqdm
```

```python
df = pd.read_csv("yelp_review.csv")
df = df[['text', 'stars']]
df = df.sample(200, random_state=42).reset_index(drop=True)
```

```python
def mock_llm(prompt):
    # placeholder to simulate LLM behaviour
    return {
        "predicted_stars": 4,
        "explanation": "Mostly positive sentiment"
    }
```

```python
correct = 0
total = len(df)

for _, row in df.iterrows():
    result = mock_llm(row["text"])
    if result["predicted_stars"] == row["stars"]:
        correct += 1

accuracy = correct / total
accuracy
```
